```python
# Adotada anÃ¡lise para buscar outliers ou pontos de atenÃ§Ã£o considerando numa primeira anÃ¡lise 2 Desvios PadrÃµes como limites (inf e sup).
```

```python
colunas = ["sales", "profit", "discount"]

for coluna in colunas:

    media = df[coluna].mean()
    desvio = df[coluna].std()

    lim_inf = media - 2 * desvio
    lim_sup = media + 2 * desvio

    outliers = df[
        (df[coluna] < lim_inf) |
        (df[coluna] > lim_sup)
    ]

    acima = (df[coluna] > lim_sup).sum()
    abaixo = (df[coluna] < lim_inf).sum()

    print("="*60)
    print(f"VariÃ¡vel: {coluna}")
    print(f"MÃ©dia: {media:.2f}")
    print(f"Desvio PadrÃ£o: {desvio:.2f}")
    print(f"Limite inferior: {lim_inf:.2f}")
    print(f"Limite superior: {lim_sup:.2f}")
    print(f"Outliers abaixo: {abaixo}")
    print(f"Outliers acima: {acima}")
    print(f"Total de outliers: {len(outliers)}")
```

```python
# Destaques observados que podem ser analisados e tratados como outliers

# Para o tratamento de dados, sugere-se inicialmente a investigaÃ§Ã£o junto as equipes comerciais, uma vez que
# podem refletir negociaÃ§Ãµes e nÃ£o necessariamente dados que devem ser expurgados.

# 247 Vendas apresentaram valores acima do limite superior indicando clientes revelantes para a empresa.
# 62 Vendas apresentaram lucro abaixo do limite inferior. Sugere-se verificar por exemplo os termos da negociaÃ§Ã£o.
# 856 Vendas apresentaram descontos acima do limite superior. Ponto de atenÃ§Ã£o dado o volume.
```

```python
# Vendas por RegiÃ£o

sales_by_region = df.groupby('region')['sales'].sum().reset_index()

print(sales_by_region)
```

```python
plt.figure(figsize=(10, 10))
plt.pie(sales_by_region['sales'], labels=sales_by_region['region'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('deep'))
plt.title('Total de Vendas por RegiÃ£o')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
